# Exploratory Data Analysis

## import libraries and load dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bike_sharing = pd.read_csv('./data/bike_sharing.csv')
```

```python
bike_sharing.head()
```

```python
bike_sharing.shape
```

## check for missing values

```python
bike_sharing.isnull().any()
```

## check outlier

```python
sns.pairplot(bike_sharing[['weathersit','temp','atemp','hum','windspeed','cnt']])
```

## convert numerical types to categorical types

```python
bike_sharing.dtypes
```

```python
bike_sharing['dteday'] = pd.to_datetime(bike_sharing.dteday)

columns_names = ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit']

for column_name in columns_names:
    bike_sharing[column_name] = bike_sharing[column_name].astype('category')

bike_sharing.dtypes
```

## categorical variables visualization

```python
plt.figure(dpi=120)
plt.title('barplot grouped by season')
season_map = {1: 'spring', 2: 'summer', 3: 'fall', 4: 'winter'}
sns.barplot(x='season', y='cnt', data=bike_sharing.assign(season=bike_sharing['season'].map(season_map)))
plt.xlabel('season')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(dpi=120)
plt.title('barplot grouped by year')
year_map = {0: '2011', 1: '2012'}
sns.barplot(x='yr', y='cnt', data=bike_sharing.assign(yr=bike_sharing['yr'].map(year_map)))
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(dpi=150)
plt.title('boxplot grouped by month')
sns.boxplot(x='mnth', y='cnt', data=bike_sharing)
plt.xlabel('month')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(dpi=150)
plt.title('boxplot grouped by hour')
sns.boxplot(x='hr', y='cnt', data=bike_sharing)
plt.xlabel('hour')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(dpi=120)
plt.title('barplot grouped by holiday')
holiday_map = {0: 'non-holiday', 1: 'holiday'}
sns.barplot(x='holiday', y='cnt', data=bike_sharing.assign(holiday=bike_sharing['holiday'].map(holiday_map)))
plt.xlabel('holiday')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(dpi=150)
plt.title('boxplot grouped by weekday')
weekday_map = {1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6:'Saturday', 0: 'Sunday'}
sns.boxplot(x='weekday', y='cnt', data=bike_sharing.assign(weekday=bike_sharing['weekday'].map(weekday_map)))
plt.xlabel('weekday')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(dpi=120)
plt.title('barplot grouped by workingday')
workingday_map = {0: 'non-workingday', 1: 'workingday'}
sns.barplot(x='workingday', y='cnt', data=bike_sharing.assign(workingday=bike_sharing['workingday'].map(workingday_map)))
plt.xlabel('workingday')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(dpi=120)
plt.title('boxplot grouped by weather condition')
weather_map = {1: 'good', 2: 'normal', 3: 'bad', 4: 'extreme'}
sns.boxplot(x='weathersit', y='cnt', data=bike_sharing.assign(weathersit=bike_sharing['weathersit'].map(weather_map)))
plt.xlabel('weather condition')
plt.ylabel('count')
plt.show()
```

## numberical variables visualization

```python
fig,axes = plt.subplots(3,2,figsize = (16,12))
sns.distplot(bike_sharing['temp'],ax=axes[0,0])
sns.distplot(bike_sharing['atemp'],ax=axes[0,1])
sns.distplot(bike_sharing['hum'],ax=axes[1,0])
sns.distplot(bike_sharing['windspeed'],ax=axes[1,1])
sns.distplot(bike_sharing['casual'],ax=axes[2,0])
sns.distplot(bike_sharing['registered'],ax=axes[2,1])
```
